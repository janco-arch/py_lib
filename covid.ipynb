```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
df = pd.read_csv("covid.csv")
df
```

```python
numOfColumns = len(df.columns)
print (f"Number Of Columns:{numOfColumns}")
```

```python
df.head(10)
```

```python
print(df.isnull().sum())
```

```python
df.dropna(subset=['date'], inplace=True)
```

```python
df['date'] = pd.to_datetime(df['date'])
```

```python
df.head(10)
```

```python
df.interpolate(method='linear', inplace=True)
```

```python
df.fillna(0, inplace=True)
```

```python
plt.plot(df["location"], df["total_cases"])# line graph plot
plt.xlabel("location")
plt.ylabel("total_cases")
plt.title("total_cases vs location")
plt.show()
```

```python
plt.plot(df["location"], df["total_deaths"])# line graph plot
plt.xlabel("location")
plt.ylabel("total_deaths")
plt.title("total_deaths vs location")
plt.show()
```

```python
df.plot(x="location", y="new_cases", kind="bar", color="blue")

plt.title("Bar Graph from new cases comparison")
plt.xlabel("location")
plt.ylabel("new_cases")
plt.show()
```

```python
df["death rate"] = df["total_deaths"].div(df["total_cases"], fill_value=1)
```
